```python
#가온지수에 콤마(,)가 포함되어 있어 계산이 불가능한 형태
#계산 가능하도록 정수형태로 바꿔주기

def to_int(가온지수):
  가온지수=가온지수.replace(',', '')
  가온지수=int(가온지수)
  return 가온지수
```

```python
가온지수_정수 = music2018_df2.가온지수.apply(to_int)
가온지수_정수.head()
```

```python
#칼럼에 추가

music2018_df2.insert(3,'가온지수_정수',가온지수_정수)
music2018_df2.head()
```

#### 상관계수 분석

상관관계는 관계를 알고자 하는 두 변수가 모두 연속변수일 경우에만 계산이 가능하고, 이는 인과관계를 분석하는게 아닌 두 연속변수가 관련이 있는지 없는지를 알아보는 기술통계 분석이다. 

두 변수를 x,y라고 하면 상관관계 계수는 x값과 y값을 모두 z값으로 표준화해서 Zx * Zy를 모두 더한다음, n(총 개수)으로 나눠준 것.

- 0 ~ 0.2 : 매우 약한 상관관계 
- 0.2 ~ 0.4 : 약한 상관관계 
- 0.4 ~ 0.6 : 어느 정도의 상관관계 
- 0.6 ~ 0.8 : 강한 상관관계 
- 0.8 ~ 1.0 : 매우 강한 상관관계 

상관계수의 범위는 -1부터 1까지. 음수가 나오면 음의 상관관계, 양수가 나오면 양의 상관관계.

numpy의 corrcoef 명령을 사용하면 상관계수를 알아서 계산해줌

```python
import numpy as np
```

```python
#-0.4 음의 상관관계를 어느전도 가짐
np.corrcoef(data_result['확진자비율'], data_result['가온지수'])
```

```python
sns.pairplot(data_result, x_vars=['확진자비율'],
             y_vars=['가온지수'], kind='reg',height=7)
plt.show()

```

####2020년만을 범위로 설정해서 그리 높은 상관관계는 보이지 않는 결과가 나와서, 2019년의 마지막 3개월 정도의 데이터도 포함하고 다시 상관계수 분석을 했더니 -0.4라는 의미있는 상관관계 결과가 드러났다. 코로나 확진자수와 음원 소비율은 반비례 관계를 꽤 가진 것으로 결론지을 수 있다.

####또한, 확진자수와는 관계없이, 가온지수는 태연, 방탄소년단, 아이유, 볼빨간사춘기 등 대형가수의 컴백에 크게 좌지우지 되는 경향이 있기 때문에 절대적인 코로나의 영향이 반영된 수치라고 이야기할 수 없다. 다만, 전체적인 영향이 존재한다는 점을 파악할 수 있었을 뿐이다.

####코로나의 등장에 따라 대중들의 음원소비율이 줄어든 까닭은 여러가지가 있을 수 있다. 드라이브나 여행 등의 신나는 상황에 음악을 곁들이는 경우가 많은데, 상황이 여의치 않다보니 무기력증 등에 의해 소비가 줄어들었을 수도 있고, 카페나 술집 등 거리에서 음악을 틀어주는 경우가 많은데 외출을 자제하는 분위기다보니 대외적으로의 소비 또한 줄어들었을 것이다.
####하지만 가장 강력한 영향은 유튜브 등 영상플랫폼의 이용률 증가가 원인이라고 생각한다.
